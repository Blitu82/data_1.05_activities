```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#1.Import the merged_clean_ver1.csv file.
df = pd.read_csv('merged_clean_ver1.csv')
df
```

```python
#2.Check the data types of all columns in the dataframe.
df.dtypes
```

```python
#3.Use the describe method on only the float type columns.
df.describe(include=["float64"]).T
```

```python
#ACTIVITY 2
#1.Get the numeric data.
numeric_df = df._get_numeric_data()
numeric_df
```

```python
#2.Use the describe() method on the numeric data in the DataFrame.
numeric_df.describe().T
```

```python
#3.Print the median.
numeric_df.median()
```

```python
#4.Print IQR.
df_iqr = numeric_df.describe().T
df_iqr
```

```python
iqr = df_iqr['75%'] - df_iqr['50%']
iqr
```

```python
#ACTIVITY 3
#1. Load data from the merged_clean_ver2.csv file.
df_2 = pd.read_csv('merged_clean_ver2.csv')
df_2
```

```python
#2. Plot Pearson Correlation Heatmap.
sns.heatmap(df_2.corr(method='pearson'), annot=True)
plt.show()
```

```python
#3. Plot Spearman Correlation Heatmap.
sns.heatmap(df_2.corr(method='spearman'), annot=True)
plt.show()
```

```python
#4. Plot Spearman Correlation Heatmap.
sns.heatmap(df_2.corr(method='kendall'), annot=True)
plt.show()
```
